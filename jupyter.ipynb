```python
from datasets import load_dataset

dataset = load_dataset("THUDM/AgentInstruct")
```

```python
data = dataset['os']['conversations']
idx = 0

human_question_split = "Now, I will start a new problem in a new OS. My problem is:"
question, answer = None, None
for i, message in enumerate(data[idx]):
    
    body = message['value']
    if human_question_split in body:
        assert message['from'] == "human"
        question = body.split(human_question_split)[1]

    if question and "Act: answer(" in body:
        assert message['from'] == "gpt"
        answer = body.split("Act: answer") [-1]
        break
        
print(question)
print()
print(answer)
```
